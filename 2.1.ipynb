```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from math import fabs
%matplotlib inline
```

```python
def get_rmse_k(x, y, k, n):
    return  sqrt(fabs(((np.mean(y ** 2) - ((np.mean(y)) ** 2)) /
                 ((np.mean(x ** 2) - 
                   ((np.mean(x)) ** 2)))) - (k ** 2))) / sqrt(n)
```

```python
def get_rmse_c(x, y, sigma_k, n):
    return sigma_k * sqrt(np.mean(x ** 2) - ((np.mean(x)) ** 2))
```

**Измерения на первой установке для измерения скорости звука при помощи раздвижной трубы**

**Экспериментальные данные:**

```python
number_of_resonanse = np.array([0, 1, 2, 3, 4])
number_of_resonanse_ = np.array([0, 1, 2, 3, 4, 5])
# длина трубы записана в метрах
delta_L_f1 = np.array([0.700, 0.755, 0.809, 0.864, 0.918])
delta_L_f2 = np.array([0.700, 0.752, 0.801, 0.852, 0.902])
delta_L_f3 = np.array([0.700, 0.748, 0.796, 0.843, 0.890])
delta_L_f4 = np.array([0.700, 0.746, 0.790, 0.834, 0.8780, 0.921])
delta_L_f5 = np.array([0.700, 0.743, 0.785, 0.826, 0.868, 0.909])
```

**Измерения для $f_{1} = 3199$ Гц**

```python
#Least squares polynomial fit. 
#Fit a polynomial p(x) = p[0] * x**deg + ... + p[deg] of degree deg to points (x, y).
#Returns a vector of coefficients p that minimises the squared error.

fig = plt.figure(figsize=(14, 9))
ax = fig.gca()
ax.set_xticks(np.arange(0, 5, 0.025), minor=True)
ax.set_yticks(np.arange(0, 1, 0.0025), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

mnk_coef_f1 = np.polyfit(number_of_resonanse, delta_L_f1, 1)
xp = np.linspace(0, 5, 100)
p = np.poly1d(mnk_coef_f1) # создаем полином с коэффициентами (убывающими по степени) из массива mnk
_ = plt.plot(number_of_resonanse, delta_L_f1, '.', xp, p(xp), '-')
plt.xlabel('$N$', horizontalalignment='right', x = 1, size = 15)
h = plt.ylabel('$L_{f1}, м$', verticalalignment='top', horizontalalignment='right', y = 1, size = 15)
h.set_rotation(0)
plt.title('$L_{f1} (N)$', size = 17)
plt.grid()
plt.scatter(number_of_resonanse, delta_L_f1, color = 'red')


plt.show()
```

**Погрешности для $f_{1}$**


```python
mnk_coef_f1
```

```python
sigma_k = get_rmse_k(number_of_resonanse, delta_L_f1, mnk_coef_f1[0], 5)
sigma_c = get_rmse_c(number_of_resonanse, delta_L_f1, sigma_k, 5) 
# среднеквадратичное отклонение для коэффициента, определенного с помощью метода наим. квадратов
```

```python
sigma_k, sigma_c
```

Так как $L = L_{0} + \frac{1}{2}\lambda n$ :

```python
lambda_half = np.array([0.0545, 0.0504, 0.0475, 0.0441, 0.0417])
lambda_ = 2 * lambda_half
frequencies = np.array([3199, 3459, 3695, 3945, 4190])
inversed_frequencies = 1 / frequencies 
```

```python
fig = plt.figure(figsize=(14, 9))
ax = fig.gca()
ax.set_xticks(np.arange(0.00023, 0.00032, 0.0000005), minor=True)
ax.set_yticks(np.arange(0.080, 0.110, 0.00025), minor=True)
ax.grid(which='minor', alpha=0.4, linestyle='-')

mnk_coef = np.polyfit(inversed_frequencies, lambda_, 1)
xp = np.linspace(0.00023, 0.00032, 100)
p = np.poly1d(mnk_coef) # создаем полином с коэффициентами (убывающими по степени) из массива mnk
_ = plt.plot(inversed_frequencies, lambda_, '.', xp, p(xp), '-')
plt.xlabel('$\\frac{1}{f}, с$', horizontalalignment='right', x = 1, size = 15)
h = plt.ylabel('$\lambda, м$',verticalalignment='top', horizontalalignment='right', y = 1, size = 15)
h.set_rotation(0)
plt.title('$\lambda (\\frac{1}{f})$', size = 17)
plt.grid()
plt.scatter(inversed_frequencies, lambda_, color = 'red')
plt.show()
```
