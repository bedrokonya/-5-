```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 6.626 * 10**(-34) #Эв*c
mu_n = 5.050 * 10**(-27) # нагугленное значение из интернета
```

```python
# функции для подсчета погрешностей
def get_rmse_k(x, y, k, n):
    return  sqrt(fabs(((np.mean(y ** 2) - ((np.mean(y)) ** 2)) /
                 ((np.mean(x ** 2) - 
                   ((np.mean(x)) ** 2)))) - (k ** 2))) / sqrt(n)

def get_rmse_c(x, y, sigma_k, n):
    return sigma_k * sqrt(np.mean(x ** 2) - ((np.mean(x)) ** 2))
```

```python
# вода 
f_0_water = np.array([9.7238,  9.0648, 9.2844, 8.7867])
B_water = np.array([228,  212, 217, 204])
```

```python
B_water
```

```python
# переводим в СИ
f_0_water = f_0_water * 10**6 
B_water = B_water * 10**(-3)
```

```python
f_0_water
```

```python
# Функция, которая высчитывает коэффициенты аппроксимирующей прямой:
#numpy.polyfit(x, y, deg, rcond=None, full=False, w=None, cov=False)
#Least squares polynomial fit. 
#Fit a polynomial p(x) = p[0] * x**deg + ... + p[deg] of degree deg to points (x, y).
#Returns a vector of coefficients p that minimises the squared error.

fig = plt.figure(figsize=(14, 9))
ax = fig.gca()
# ax.set_xticks(np.arange(0, 5, 0.05), minor=True) # это для мелких делений по оси ОХ
# ax.set_yticks(np.arange(0, 1, 0.005), minor=True) # аналогично для OY
ax.grid(which='minor', alpha=0.4, linestyle='-')

mnk_coef_f1 = np.polyfit(B_water, f_0_water, 1) # считаем коэфф-ты аппроксимирующей прямой
xp = np.linspace(0.200, 0.230, 100)
p = np.poly1d(mnk_coef_f1) # создаем полином с коэффициентами (убывающими по степени) из массива mnk_coef_f1
_ = plt.plot(B_water, f_0_water, '.', xp, p(xp), '-') # рисуем точки и аппроксимирующую прямую
plt.xlabel('$B$, мТл', horizontalalignment='right', x = 1, size = 15)
h = plt.ylabel('$f$, МГц ', verticalalignment='top', 
               horizontalalignment='right', y = 1, size = 15)
h.set_rotation(0)
plt.title('$f(B)$', size = 17)
plt.grid()
plt.scatter(B_water, f_0_water, color = 'red')


plt.show()
```

```python
#резина
f_0_rubber = np.array([7.4649, 7.2857, 6.9371, 6.6360, 6.6523])
B_rubber = np.array([174.6, 170.8, 162.6, 154.1, 151.2])
```

```python
B_rubber
```

```python
f_0_rubber = f_0_rubber * 10**6
B_rubber = B_rubber * 10**(-3)
```

```python
# Функция, которая высчитывает коэффициенты аппроксимирующей прямой:
#numpy.polyfit(x, y, deg, rcond=None, full=False, w=None, cov=False)
#Least squares polynomial fit. 
#Fit a polynomial p(x) = p[0] * x**deg + ... + p[deg] of degree deg to points (x, y).
#Returns a vector of coefficients p that minimises the squared error.

fig = plt.figure(figsize=(14, 9))
ax = fig.gca()
ax.set_xticks(np.arange(0, 5, 0.05), minor=True) # это для мелких делений по оси ОХ
ax.set_yticks(np.arange(0, 1, 0.005), minor=True) # аналогично для OY
ax.grid(which='minor', alpha=0.4, linestyle='-')

mnk_coef_f1 = np.polyfit(B_rubber, f_0_rubber, 1) # считаем коэфф-ты аппроксимирующей прямой
xp = np.linspace(0.1500, 0.1800, 100)
p = np.poly1d(mnk_coef_f1) # создаем полином с коэффициентами (убывающими по степени) из массива mnk_coef_f1
_ = plt.plot(B_rubber, f_0_rubber, '.', xp, p(xp), '-') # рисуем точки и аппроксимирующую прямую
plt.xlabel('$B$, мТл', horizontalalignment='right', x = 1, size = 15)
h = plt.ylabel('$f$, МГц ', verticalalignment='top', 
               horizontalalignment='right', y = 1, size = 15)
h.set_rotation(0)
plt.title('$f(B)$', size = 17)
plt.grid()
plt.scatter(B_rubber, f_0_rubber, color = 'red')


plt.show()
```
